```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Create a bar chart of the average arrival times 
fig, ax = plt.subplots(figsize = (20,20))

# Label plot
ax.set_title('Class Arrival Times')

ax.set_xlabel('Days')
ax.set_ylabel('Average Arrival Time')

# Generate data
x = (class_data_df.columns)
y = (class_data_df['Day_1'].mean(),class_data_df['Day_2'].mean(),class_data_df['Day_3'].mean(),
     class_data_df['Day_4'].mean(),class_data_df['Day_5'].mean(),class_data_df['Day_6'].mean())

# Plot
ax.barh(x, y, color="blue")
```
